```python
country_respodents = df['Country'].value_counts()
country_respodents
```

```python
country_use_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_use_python
```

```python
python_df = pd.concat([country_respodents, country_use_python], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'LanguageWorkedWith': 'NumKnowsPython', 'Country': 'NumberRespondents'}, inplace=True)
```

```python
python_df
```

```python
python_df['PctKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumberRespondents']) * 100
python_df
```

```python
python_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True)
```
