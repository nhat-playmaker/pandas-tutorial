```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'],
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'],
    'email': ['CoreySchafer@gmail.com', 'JaneDoe@gmail.com', 'JohnDoe@gmail.com', None, np.nan, 'Anonymus@gmail.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df = pd.DataFrame(people)
df.replace('NA', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)
```

```python
df
```

```python
# Drop NaN
df.dropna()

# default: how='any'
# Means any rows have NaN or None -> drop
```

```python
df.dropna(axis='index', how='all')
# how = 'any' --> any row has NaN value -> drop
# how = 'all' --> all columns of the row are NaN -> drop

# Cuz line 4 is all NaN and None -> drop
```

```python
df.dropna(axis='index', how='all', subset=['email'])
# if email is NaN or None -> drop
```

```python
df.dropna(axis='index', how='all', subset=['email', 'last'])
# if email AND last are NaN or None -> drop
```

```python
df.dropna(axis='index', how='any', subset=['email', 'last'])
# if email OR last are NaN or None -> drop
```

```python
df.isna()
```

```python
df.fillna(0)
# Fill all NaN values
```
